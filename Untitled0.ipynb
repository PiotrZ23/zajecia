```python
# === GŁÓWNY SKRYPT APLIKACJI ===
from flask import Flask, request, redirect, render_template_string
from flask_ngrok import run_with_ngrok
import sqlite3

app = Flask(__name__)
run_with_ngrok(app)

# === Tworzenie bazy danych ===
def init_db():
    conn = sqlite3.connect("pacjenci.db")
    c = conn.cursor()
    c.execute("""
        CREATE TABLE IF NOT EXISTS pacjenci (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            imie TEXT,
            nazwisko TEXT,
            email TEXT UNIQUE,
            haslo TEXT,
            pesel TEXT,
            wiek INTEGER
        )
    """)
    conn.commit()
    conn.close()

init_db()

# === Strona główna z rejestracją ===
@app.route("/", methods=["GET", "POST"])
def rejestracja():
    if request.method == "POST":
        dane = (
            request.form["imie"],
            request.form["nazwisko"],
            request.form["email"],
            request.form["haslo"],
            request.form["pesel"],
            request.form["wiek"]
        )

        try:
            conn = sqlite3.connect("pacjenci.db")
            c = conn.cursor()
            c.execute("INSERT INTO pacjenci (imie, nazwisko, email, haslo, pesel, wiek) VALUES (?, ?, ?, ?, ?, ?)", dane)
            conn.commit()
            conn.close()
            return redirect("/login")
        except:
            return "Błąd: Ten email już istnieje w bazie danych."

    return render_template_string("""
    <h2>Rejestracja pacjenta</h2>
    <form method="POST">
        Imię: <input name="imie"><br>
        Nazwisko: <input name="nazwisko"><br>
        Email: <input name="email" type="email"><br>
        Hasło: <input name="haslo" type="password"><br>
        PESEL: <input name="pesel"><br>
        Wiek: <input name="wiek" type="number"><br>
        <input type="submit" value="Zarejestruj">
    </form>
    """)

# === Logowanie pacjenta ===
@app.route("/login", methods=["GET", "POST"])
def login():
    if request.method == "POST":
        email = request.form["email"]
        haslo = request.form["haslo"]

        conn = sqlite3.connect("pacjenci.db")
        c = conn.cursor()
        c.execute("SELECT * FROM pacjenci WHERE email=? AND haslo=?", (email, haslo))
        user = c.fetchone()
        conn.close()

        if user:
            return redirect(f"/pacjent/{user[0]}")
        else:
            return "Nieprawidłowy email lub hasło."

    return render_template_string("""
    <h2>Logowanie</h2>
    <form method="POST">
        Email: <input name="email" type="email"><br>
        Hasło: <input name="haslo" type="password"><br>
        <input type="submit" value="Zaloguj się">
    </form>
    """)

# === Karta pacjenta ===
@app.route("/pacjent/<int:id>")
def pacjent(id):
    conn = sqlite3.connect("pacjenci.db")
    c = conn.cursor()
    c.execute("SELECT * FROM pacjenci WHERE id=?", (id,))
    pacjent = c.fetchone()
    conn.close()

    if not pacjent:
        return "Nie znaleziono pacjenta."

    return render_template_string(f"""
    <h2>Karta pacjenta</h2>
    <ul>
        <li>Imię: {pacjent[1]}</li>
        <li>Nazwisko: {pacjent[2]}</li>
        <li>Email: {pacjent[3]}</li>
        <li>PESEL: {pacjent[5]}</li>
        <li>Wiek: {pacjent[6]}</li>
    </ul>
    <a href="/login">Wyloguj</a>
    """)

# === Uruchom aplikację ===
app.run()
```
